```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

def fetch_btcusd_data():
    btc = yf.Ticker("BTC-USD")
    df = btc.history(period="1y")  # Fetch 1 year of daily data
    return df[['Close']]

def calculate_stock_span(prices):
    n = len(prices)
    span = [0] * n  # Span array
    stack = []  # Stack to store index

    for i in range(n):
        while stack and prices[i] >= prices[stack[-1]]:
            stack.pop()

        span[i] = i + 1 if not stack else i - stack[-1]
        stack.append(i)

    return span

def analyze_and_plot(df):
    prices = df['Close'].values
    span = calculate_stock_span(prices)
    df['StockSpan'] = span

    plt.figure(figsize=(12, 6))

    # Plot closing price
    plt.subplot(2,1,1)
    plt.plot(df.index, df['Close'], label='BTC-USD Close Price', color='blue')
    plt.title('BTC-USD Price and Stock Span')
    plt.legend()

    # Plot stock span
    plt.subplot(2,1,2)
    plt.plot(df.index, df['StockSpan'], label='Stock Span', color='red')
    plt.xlabel('Date')
    plt.legend()

    plt.show()

if __name__ == "__main__":
    df = fetch_btcusd_data()
    analyze_and_plot(df)
```

```python
# Task2
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import bisect

class SegmentTree:
    def __init__(self, data):
        self.n = len(data)
        self.tree_min = [0] * (4 * self.n)
        self.tree_max = [0] * (4 * self.n)
        self.data = data
        self.build(0, 0, self.n - 1)

    def build(self, node, start, end):
        if start == end:
            self.tree_min[node] = self.data[start]
            self.tree_max[node] = self.data[start]
        else:
            mid = (start + end) // 2
            left_child = 2 * node + 1
            right_child = 2 * node + 2
            self.build(left_child, start, mid)
            self.build(right_child, mid + 1, end)
            self.tree_min[node] = min(self.tree_min[left_child], self.tree_min[right_child])
            self.tree_max[node] = max(self.tree_max[left_child], self.tree_max[right_child])

    def query_min(self, node, start, end, L, R):
        if L > end or R < start:
            return float('inf')
        if L <= start and end <= R:
            return self.tree_min[node]
        mid = (start + end) // 2
        left_min = self.query_min(2 * node + 1, start, mid, L, R)
        right_min = self.query_min(2 * node + 2, mid + 1, end, L, R)
        return min(left_min, right_min)

    def query_max(self, node, start, end, L, R):
        if L > end or R < start:
            return float('-inf')
        if L <= start and end <= R:
            return self.tree_max[node]
        mid = (start + end) // 2
        left_max = self.query_max(2 * node + 1, start, mid, L, R)
        right_max = self.query_max(2 * node + 2, mid + 1, end, L, R)
        return max(left_max, right_max)

    def update(self, index, value, node, start, end):
        if start == end:
            self.tree_min[node] = value
            self.tree_max[node] = value
        else:
            mid = (start + end) // 2
            if index <= mid:
                self.update(index, value, 2 * node + 1, start, mid)
            else:
                self.update(index, value, 2 * node + 2, mid + 1, end)
            self.tree_min[node] = min(self.tree_min[2 * node + 1], self.tree_min[2 * node + 2])
            self.tree_max[node] = max(self.tree_max[2 * node + 1], self.tree_max[2 * node + 2])

def fetch_btcusd_data():
    btc = yf.Ticker("BTC-USD")
    df = btc.history(period="1y")  # Fetch 1 year of daily data
    return df[['Close']]

def analyze_and_plot(df, seg_tree):
    prices = df['Close'].values
    df['Support'] = [seg_tree.query_min(0, 0, len(prices) - 1, max(0, i - 10), i) for i in range(len(prices))]
    df['Resistance'] = [seg_tree.query_max(0, 0, len(prices) - 1, max(0, i - 10), i) for i in range(len(prices))]

    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['Close'], label='BTC-USD Close Price', color='blue')
    plt.plot(df.index, df['Support'], label='Support Level', linestyle='dashed', color='green')
    plt.plot(df.index, df['Resistance'], label='Resistance Level', linestyle='dashed', color='red')
    plt.title('BTC-USD Price with Support & Resistance Levels')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    df = fetch_btcusd_data()
    seg_tree = SegmentTree(df['Close'].values.tolist())
    analyze_and_plot(df, seg_tree)
```
